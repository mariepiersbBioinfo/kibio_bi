## Catenate 11 RNA dataframes into one

```python
rna = pd.concat([rna1, rna2, rna3, rna4, rna5, rna6, rna7, rna8, rna9, rna10, rna11])
print("rna", rna.shape)
```

## Catenate the 4 main datasets

```python
observations = pd.concat([rna, clinical, metabolite, gut])
print("observations", observations.shape)
#observations
```

## Get list of unique samples

```python
sample_id = ed.eland_to_pandas(ed.DataFrame(es_client= es, es_index_pattern='multiomics_sampleid_type'))
print("sample_id", sample_id.shape)
sample_id.drop(['@columns','id','type','IRIS','BMI','BMI_category'], axis=1, inplace=True)
print("sample_id", sample_id.shape)
sample_id = sample_id.drop_duplicates()
print("sample_id", sample_id.shape)
```

## Merge observations with sample_id and produce final dataframe

```python
observations_sample = pd.merge(observations, sample_id, on="SampleID")
print("observations_sample", observations_sample.shape)
#observations_sample
```
